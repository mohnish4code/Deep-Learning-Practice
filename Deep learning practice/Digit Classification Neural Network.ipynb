```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

**Flatten the array**

```python
X_train.shape
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train_flated = X_train.reshape(len(X_train), 28*28)
X_train_flated
```

```python
X_test_flated = X_test.reshape(len(X_test), 28*28)
X_test_flated
```

```python
X_train_flated[0]
```

```python
X_test.shape
```

```python
X_test_flated.shape
```

**Creation of the simple neural network**

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape = (784,),activation="sigmoid")
])

model.compile(optimizer="adam",
              loss = "sparse_categorical_crossentropy",
              metrics = ["accuracy"]
)

model.fit(X_train_flated, y_train, epochs=7)
```

```python
model.evaluate(X_test_flated, y_test)
```

```python
plt.matshow(X_test[0])
```

```python
y_predicted = model.predict(X_test_flated)
y_predicted[0]
```

```python
np.argmax(y_predicted[0])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
```

```python
cm = tf.math.confusion_matrix(labels = y_test, predictions = y_predicted_labels)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth values")
```

**Hidden layer in the model**

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape = (784,),activation="relu"),
    keras.layers.Dense(10, activation="sigmoid")
])

model.compile(optimizer="adam",
              loss = "sparse_categorical_crossentropy",
              metrics = ["accuracy"]
)

model.fit(X_train_flated, y_train, epochs=7)
```

```python
model.evaluate(X_test_flated, y_test)
```

```python
y_predicted = model.predict(X_test_flated)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels = y_test, predictions = y_predicted_labels)
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt="d")
plt.xlabel("Predicted")
plt.ylabel("Truth values")
```
