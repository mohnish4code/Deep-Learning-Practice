```python
import numpy as np
y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

**For mean absolute error**

```python
def mae(y_predicted, y_true):
    total_error = 0
    n = len(y_true)
    for yp, yt in zip(y_predicted, y_true):
        total_error += abs(yp-yt)
    mae = total_error/n
    return f"Mean absolute error is : {mae} and Total error is : {total_error}"
```

```python
mae(y_predicted, y_true)
```

```python
def mae(y_predicted, y_true):
    total_error = np.sum(np.abs(y_predicted - y_true))
    mae = np.mean(np.abs(y_predicted - y_true))
    return f"Total error is {total_error} and mean_absolute_error is {mae}"
```

```python
mae(y_predicted, y_true)
```

```python
def log_loss(y_true, y_predicted):
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
log_loss(y_true, y_predicted)
```

**Mean squared error**

```python
import math
def mse(y_predicted, y_true):
    total_error = 0
    n = len(y_true)
    for yp, yt in zip(y_predicted, y_true):
        total_error += math.pow((yp-yt), 2)
    mse = total_error/n
    return f"Total error is {total_error} and mean_squared_error is {mse}"
```

```python
mse(y_predicted, y_true)
```

**Now using the numpy**

```python
def mse(y_predicted, y_true):
    total_error = np.sum(np.square(y_predicted-y_true))
    mse = np.mean(np.square(y_predicted-y_true))
    return f"Total error is {total_error} and mean_squared_error is {mse}"
```

```python
mse(y_predicted, y_true)
```
