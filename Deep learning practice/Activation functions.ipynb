```python
import math
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
sigmoid(-2)
```

```python
sigmoid(10)
```

**This is for the sigmoid function**

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(4)
```

```python
tanh(-23)
```

**Tanh Function**

```python
def relu(x):
    return max(0, x)
```

```python
relu(-9)
```

```python
relu(32)
```

**ReLU function**

```python
def LeakyReLU(x):
    return max(0.1*x, x)
```

```python
LeakyReLU(23)
```

```python
LeakyReLU(-12)
```
